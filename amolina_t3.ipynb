```python
#Bibliotecas
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

#Datos
years = [1975, 1993, 2018]
year = [2010]
dato = [4]
regiones = {
    "Antsiranana":[ .597982, .954733, 2.012749],
    "Mahajanga":[ .819750, 1.364793, 3.135764],
    "Toamasina":[ 1.179606, 1.995461, 3.892262],
    "Antananarivo":[ 2.167973, 3.601127, 7.264922],
    "Fianarantsoa":[ 1.804365, 2.550190, 5.162291],
    "Toliary":[ 1.034114, 1.772610, 4.212354],
}

#Grafico principal
fig, ax = plt.subplots()

#Barras
x = np.arange(len(years))
width = 0.15

for i, (region, populations) in enumerate(regiones.items()):
    rects = ax.bar(x + i*width, populations, width, label=region)

#Ejes
ax.set_ylabel("población (millones)")
ax.set_title("Población de Madagascar")
ax.set_xticks(x + width * (len(regiones) - 1) / 2)
ax.set_xticklabels(years)

#Leyenda
ax.legend(loc='upper left', bbox_to_anchor=(1, 1))

#Grafico pequeño
ax_inset = fig.add_axes([0.20, 0.55, 0.25, 0.25])

ax_inset.bar(year, dato)
ax_inset.set_title('Pingüinos')
ax_inset.set_ylabel("cantidad")
ax_inset.tick_params(axis='x', labelsize=8)
ax_inset.tick_params(axis='y', labelsize=8)
ax_inset.set_xticks(years)

#Mostrar gráfico
plt.show()
```

##Provincias
Madagascar durante mucho tiempo se separó en 6 grandes provincias. Estas provincias fueron:

  1. Antsiranana
  2. Mahajanga
  3. Toamasina
  4. Antananarivo
  5. Fianarantsoa
  6. Toliary

###Datos
En el año 2004 las provincias e se disolvieron para formar 22 regiones, donde posteriormente se crearon dos mas para lograr un total de 24 regiones. Pero para efectos de estudio de población se utilizaron las 6 provincias antes de su disolución.
